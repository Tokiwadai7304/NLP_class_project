```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.model_selection import train_test_split

```

```python
df = pd.read_csv("./final data.csv")
df.head()
```

```python
x = df[["Description"]]
y = df[["Label"]]
x
```

```python
y
```

```python
# Khởi tạo TfidfVectorizer
vectorizer = TfidfVectorizer()

# Chuyển đổi dữ liệu văn bản thành ma trận TF-IDF
x_tfidf = vectorizer.fit_transform(x["Description"])  # Chuyển đổi x

```

```python
print(x_tfidf.toarray())
```

```python
y_class = []
for label in y["Label"]:
    if label == "truyện":
        label = 3
    elif label == "giáo dục":
        label = 0
    elif label == "phim":
        label = 1
    else:
        label = 2
    y_class.append([label]) # Thêm label vào y_class

class_name = ["giáo dục", "phim", "thể thao", "truyện" ]
print(y_class)
```

```python
print(y_class)
```

```python

x_train, x_test, y_train, y_test = train_test_split(x_tfidf, y_class, test_size=0.2, random_state=42, shuffle=True)


print(x_train)
print(y_train)
```
